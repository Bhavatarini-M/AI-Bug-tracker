```python
#approach 1
def div(a,b):
    if(b==0):
        print("zero div not possible")
        return
    else:
        return a/b
print(div(5,2))
print(div(2,2))
print(div(2,6))
print(div(2,0))
print(div(20,2))
print(div(2,2))
print(div(2,6))
```

```python
#approach 2
def update_div(fun):
    def update(a,b):
        if(b==0):
            print("zero div not possible")   
        else:
            return fun(a,b)
    return update
def div(a,b):
    return a/b
k=update_div(div)
print(k(5,2))
print(k(2,2))
print(k(2,6))
print(k(2,0))
print(k(20,2))
print(k(2,2))
print(k(2,6))
```

```python
def update_div(fun):
    def update(a,b):
        if(b==0):
            print("zero div not possible")   
        else:
            return fun(a,b)
    return update
@update_div
def div(a,b):
    return a/b
print(k(5,2))
print(k(2,2))
print(k(2,6))
print(k(2,0))
print(k(20,2))
print(k(2,2))
print(k(2,6))
```

```python
def update_div(fun):
    def update(a,b):
        if(b==0):
            print("zero div not possible")   
        else:
            return fun(a,b)
    return update
def div(a,b):
    return a/b
k=update_div(div)
print(k(5,2))
print(k(2,2))
print(k(2,6))
print(k(2,0))
print(k(20,2))
print(k(2,2))
print(k(2,6))
```

```python
def update2(fun):
    def update_greet(name):
        if(name=="Bizu"):
            print(f"hi {name} sir")
        else:
            return fun(name)
    return update_greet
def update1(fun):
    def update_greet(name):
        if(name=="Shanthi"):
            print(f"hi {name} maam")
        else:
            return fun(name)
    return update_greet
@update2
@update1
def greet(name):
    print(f"hi {name}")
greet("Sanjay")
greet("Srini")
greet("Shanthi")
greet("Bizu")
```

```python
def AddBreadLayer(fun):
    def wrapper():
        print("Bread1")
        fun()
        print("Bread2")
    return wrapper
def AddCheeseLayer(fun):
    def wrapper():
        print("Cheese1")
        fun()
        print("Cheese2")
    return wrapper
@AddBreadLayer
@AddCheeseLayer
def potatostuff():
    print("Potato")
potatostuff()
```

```python
def checkLog(fun):
    def wrapper(*args):
        start = time.time()
        fun(*args)
        end = time.time()
        print(end-start)
    return wrapper
```

```python
@checkLog
def add(a,b):
    print("add_func")
add(10,20)
```

```python
@checkLog
def slow_fun():
    time.sleep(5)
    print("slow_fun")
slow_fun()
```
