```python
#TASK SCHEDULER USING deque and headq
from collections import deque  #deque -- double ended queue
import heapq

#Create empty structures
tasks = deque()
priority_tasks = []

#Add tasks to queue
tasks.append((3, "Backup"))
tasks.append((1, "Email"))
tasks.append((2, "Analytics"))

#Move from queue to heap
while tasks:
    heapq.heappush(priority_tasks, tasks.popleft())

#Run tasks in order of priority
while priority_tasks:
    priority, name = heapq.heappop(priority_tasks)
    print(f"Running {name} (Priority: {priority})")

print("All tasks completed!")
```
