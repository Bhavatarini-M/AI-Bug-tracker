**Student Database System**

```python
import sqlite3

#-----------------Database Setup--------------------
def connect_db():
    conn = sqlite3.connect("students.db")
    return conn
#-----------------CRUD Fucntions--------------------
def create_table():
    conn = connect_db()
    cur = conn.cursor()
    cur.execute('''CREATE TABLE IF NOT EXISTS students(
                    id INTEGER PRIMARY KEY AUTOINCREMENT,
                    name TEXT,
                    age INTEGER,
                    grade TEXT)''')
    conn.commit()
    conn.close()
    print("Table created successfully.")

def insert_student(name, age, grade):
    conn = connect_db()
    cur = conn.cursor()
    cur.execute("INSERT INTO students (name, age, grade) VALUES (?, ?, ?)",
                (name, age, grade))
    conn.commit()
    conn.close()
    print(f"Student added: {name}, Age: {age}, Grade: {grade}")

def view_students():
    conn = connect_db()
    cur = conn.cursor()
    cur.execute("SELECT * FROM students")
    rows = cur.fetchall()
    conn.close()
    print("All Students:")
    for row in rows:
        print(row)

def update_students(student_id, grade):
    conn = connect_db()
    cur = conn.cursor()
    cur.execute("UPDATE students SET grade = ? WHERE id = ?",
                (grade, student_id))
    conn.commit()
    conn.close()
    print(f"Updated student ID {student_id} to Grade {grade}")

def delete_student(student_id):
    conn = connect_db()
    cur = conn.cursor()
    cur.execute("DELETE FROM students WHERE id = ?", (student_id,))
    conn.commit()
    conn.close()
    print(f"Deleted student ID {student_id}")
    
#-----------------------------Main Execution-----------------------------------
if __name__ == "__main__":
    create_table()
    insert_student("ALice", 20, "A")
    insert_student("Bob", 21, "B")
    view_students()
    update_students(1, "A+")
    delete_student(2)
    view_students()
    print("Program executed successfully.")
    
```

try: Contains code that might raise an excception.
except: Catches and handles the exception.
else: Runs only if no exception occurs.
finally: Runs always (cleanup code).
    

```python
try:
    y = 10 / 0
    print("hi try")
except:
    print("Please enter a valid number") # if except is executed else part is not print 
else:
    print("No error occured.")
finally:
    print("Program ended.")
    
```

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    lst = [1, 2, 3]
    print(lst[num])
except ZeroDivisionError:
    print("You cannot divide by zero.")
except IndexError:
    print("Index out of range.")
except Exception as e:
    print("Some other error occured.")
else:
    print("No exception occured.")
finally:
    print("Program finished.")
```

```python
try:
     print("Outer try block start")
     try:
         x = int(input("Enter a number: "))
         y = 10 / x
         print("Result: ",y)
     except ValueError:
         print("Inner block: Division by zero error.")
     finally:
         print("Inner finally executed.")
except ZeroDivisionError: 
    print("Outer block: Invalid input (not a number).")
finally:
     print("Outer finally executed.")
```

```python
#Example 2: Raising Inside try except
try:
    raise ZeroDivisionError("Manual division error!")
except ZeroDivisionError as e:
    print("Caught:",e)
#the raised error is caught and handled.
```

```python
#Example 3: Raising 
class AgeError(Exception):
    """Custom exception for invalid age."""
    pass
age = int(input("Enter age: "))
if age < 18:
    raise AgeError("You must be 18 or older!")
```

```python
#Example 4: Re-raising on Exception
try:
    x = 10 / 0
except ZeroDivisionError as e:
    print("Logging error:", e)
    raise     #re-raises the same exception
```

```python
class A:
    def __init__(self):
        self.mylist = {"Jack":["C"],"Bala":["Python","C","Java","C++"]}
        print("hello guys")
    def update(self):
        name = input()
        skill = input()
        self.mylist[name].append(skill)
    def view(self):
        print(f"slist: {self.mylist}")
obj = A()
#obj.initialization() #if this init is given the def will change into initialization instead of __init__ 
obj.update()   #if __init__ is given in the def the obj.initialization is not used (it will be used when first def is changed into initialization)
obj.view()
```

**from abc import ABC, abstractmethod**

```python
from abc import ABC, abstractmethod  #--> Parent class is not implement ,child class is implement when using the abstractmethod
class Parent(ABC):
    @abstractmethod
    def marraige(self):
        pass  #to delay using the pass (we don't left any part like empty in python so we use the pass)
class Child(Parent):
    def marraige(self):
        print("Kathija")
obj = Child()
obj.marraige()
```

```python
#Aggregation (Weak "has-a" relationship)
#The contained object can exist independently of the container.
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = engine  #Engine is passed in (aggregation)

    def drive(self):
        self.engine.start()
        print("Car is moving")

eng = Engine()
car = Car(eng)
car.drive()
'''
Engine exists outside the Car.
Even if Car is destroyed, Engine still exists.
Loose coupling.
'''
```
