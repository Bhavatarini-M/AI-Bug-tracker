```python
def div(a,b):
    return a/b
print(div(5,2))
print(div(2,2))
print(div(2,6))
print(div(2,0)) #if the denominator is zero the func will not run after that and itself
print(div(20,2))
print(div(2,2))
print(div(2,6))
```

```python
#approach 1
def div(a,b):
    if(b==0):
        print("zero div not possible")
        return
    else:
        return a/b
print(div(5,2))
print(div(2,2))
print(div(2,6))
print(div(2,0))
print(div(20,2))
print(div(2,2))
print(div(2,6))
```

```python
#approach 2
def update_div(fun):
    def update(a,b):
        if(b==0):
            print("zero div not possible")   
        else:
            return fun(a,b)
    return update
def div(a,b):
    return a/b
k=update_div(div)
print(k(5,2))
print(k(2,2))
print(k(2,6))
print(k(2,0))
print(k(20,2))
print(k(2,2))
print(k(2,6))
```
